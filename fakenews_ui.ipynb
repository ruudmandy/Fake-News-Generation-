```python
from ipywidgets import widgets, interact
from IPython.display import display, clear_output
from transformers import MT5ForConditionalGeneration, T5Tokenizer
import tweepy
from termcolor import colored
```

```python
# Change Model Name befor generate
MODEL_NAME = 'nonamenlp/news_gen' 
TOKENIZER_NAME = "nonamenlp/news_gen"
trained_model = MT5ForConditionalGeneration.from_pretrained(MODEL_NAME, return_dict=True)
tokenizer = T5Tokenizer.from_pretrained(TOKENIZER_NAME)
```

```python
def generate_answer(tags):
    source_encoding = tokenizer(
        tags,
        max_length = 150,
        padding = 'max_length',
        truncation = True,
        return_attention_mask = True,
        add_special_tokens = True,
        return_tensors='pt'
    )
    generated_ids = trained_model.generate(
        input_ids=source_encoding['input_ids'],
        attention_mask=source_encoding['attention_mask'],
        num_beams=5,
        max_length=150,
        repetition_penalty=2.5,
        length_penalty=1.0,
        early_stopping=True,
        use_cache=True
    )

    preds = [
             tokenizer.decode(generated_id, skip_special_tokens=True, clean_up_tokenization_spaces=True)
             for generated_id in generated_ids
    ]
    return "".join(preds)
```

```python
#display(button, output)
def on_button_clicked_generate(b):
    global gen_text
    with output:
        clear_output()
        keylist = [key1.value, key2.value, key3.value, key4.value, key5.value]
        keywords = " ".join(keylist)
        print("\n Yours keywords is {}".format(keylist))
        gen_text = generate_answer(keywords)
        print('\n The generate News is :')
        print(gen_text)


def on_button_clicked_twitter(b):
        if len(gen_text) < 1:
           print(colored('Please generate fake news before post on twiiter.', 'red'))
        else:
         CUSTOMER_KEY = ""
         CUSTOMER_SECRET_KEY = ""
         ACCESS_TOKEN_KEY = ""
         ACCESS_TOKEN_SECRET_KEY = ""

         # Authenticate to Twitter
         auth = tweepy.OAuthHandler(CUSTOMER_KEY, CUSTOMER_SECRET_KEY)
         auth.set_access_token(ACCESS_TOKEN_KEY, ACCESS_TOKEN_SECRET_KEY)

         # Create API object
         api = tweepy.API(auth)

         # Create a tweet
         api.update_status(gen_text)
         print('fake news was poested on https://twitter.com/Pchomp1')
```

##Demo

```python
#@title #Fake News Generator
key1 = widgets.Text(placeholder="Type something", description="Keyword1: ", disabled=False)
key2 = widgets.Text(placeholder="Type something", description="Keyword2: ", disabled=False)
key3 = widgets.Text(placeholder="Type something", description="Keyword3: ", disabled=False)
key4 = widgets.Text(placeholder="Type something", description="Keyword4: ", disabled=False)
key5 = widgets.Text(placeholder="Type something", description="Keyword5: ", disabled=False)
button = widgets.Button(description="Generate")
button2 = widgets.Button(description="Post on Twitter")
output = widgets.Output()

button.on_click(on_button_clicked_generate)
button2.on_click(on_button_clicked_twitter)

gen_text = ''

print("Please define keywords for fake news generation")
display(key1)
display(key2)
display(key3)
display(key4)
display(key5)
display(button, output)
#display(button2)
```
